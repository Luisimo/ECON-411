### 4. Import a data set for your project in pandas. You may use whatever means you prefer to import the data. Print the dataframe.

```python
filename = "POILBREUSDM.csv"
proj_df = pd.read_csv(filename)
proj_df
```

```python
adj_proj_df = proj_df[::12].set_index("DATE")
adj_proj_df
```

### 5. Create a new column of data derived from the data that you have already imported. Hint: this may be a ratio or whatever trasnforamtion that you think would be useful.

```python
adj_proj_df["Difference"] = adj_proj_df["POILBREUSDM"].diff()
adj_proj_df
```

### 6. Use a for loop to print individually each column in the data frame.

```python
for column in adj_proj_df:
    column_data = adj_proj_df[column]
    print("Column Name:", column)
    print(column_data.values)
```

### 7. Create a function that plots 2 to 3 differenct variables from the dataframe. The function may look something like "def plot_data(df,names,etc...). Each plot should include a legend and title.

```python
adj_proj_df.plot(secondary_y = "Difference")
```

### 8. Create a bar graph plot using the data you have imported

```python
adj_proj_df.plot.bar(secondary_y = "Difference")
```

### 9. On any of the Graphs you have just created, change the font size, axis labels, tick laels, color, linewidth, and so forth. Adjust these parameters to make the visualization aesthetically appealing.

```python
my_colors = ["C1", "C2"]
fig, ax = plt.subplots(figsize = (12,8))
adj_proj_df.plot(secondary_y = "Difference",
                linewidth = 4,
                alpha = 0.75,
                fontsize = 15,
                color = my_colors,
                ax = ax)
y1_ticklabels = ["$" + str(y) for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)
```
