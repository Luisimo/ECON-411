### 2. Use the variance function from the textbook to find the variance of this list. Assume that the list represent a population in whole.

```python
# total function from textbook:
def total(lst):
    total_ = 0
    n = len(lst)
    for i in range(n):
        total_ += lst[i]
    return total_
```

```python
# mean function from textbook:
def mean(lst):
    n = len(lst)
    mean_ = total(lst) / n
    return mean_
```

```python
# variance function from textbook:
def variance(lst, sample = True):
    list_mean = mean(lst)
    n = len(lst)
    DoF = n - 1
    sum_sq_diff = 0
    for val in lst:
        sum_sq_diff += (val - list_mean) ** 2
    if sample == False:
        variance_ = sum_sq_diff / n
    else:
        variance_ = sum_sq_diff / DoF
    
    return variance_

x1_var = variance(x1, sample = False)
x1_var
```

### 5. At the end of chapter 2, we used for loops to find min and max values. Create a min() function and max() function and pass the values from the list in question 1 to each of these to determine the min and max values in the list.

```python
test_list = [6,7,8]

# named the function "minn()" because "min()" already exists and I wanted to
# avoid possible overriding issues.
def minn(lst):
    min_list_val = float("inf")
    max_list_val = float("-inf")
    for x in lst:
        if x < min_list_val:
            min_list_val = x
#        if x > max_list_val:
#            max_list_val = x
    return min_list_val

minn(test_list)
```

```python
# named the function "maxx()" because "max()" already exists and I wanted to
# avoid possible overriding issues.
def maxx(lst):
    min_list_val = float("inf")
    max_list_val = float("-inf")
    for x in lst:
#        if x < min_list_val:
#            min_list_val = x
        if x > max_list_val:
            max_list_val = x
    return max_list_val

maxx(test_list)
```

### 6. Explain why it might be advantageous to create a function instead of building all commands from scratch as you create a script.

If you create a function of a command that you know you will use often, it will save you a lot of time later on in your code. Sure, you could copy and paste the command if you need to use it multiple times, but then you will need to go through your code and adjust the values each time. Depending on how low your command is, this process could be quite tedious. For more complex projects, creating your own functions could save you some time in the long run.
