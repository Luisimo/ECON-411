```python
import pandas as pd
import datetime
import pandas_datareader.data as web
import matplotlib.pyplot as plt
```

```python
def gather_data(data_codes,
               start,end = datetime.datetime.today(),
               freq = "M"):
    df_init = False
    for key, code in data_codes.items():
        if df_init == False:
            df = web.DataReader(code, "fred", start, end).resample(freq).mean() 
            df.rename(columns = {code:key}, inplace = True)
            df_init = True
        else:
            df[key] = web.DataReader(code, "fred", start, end).resample(freq).mean()
            
    return df
```

```python
data_codes = {"Price Per Barrel": "DCOILBRENTEU",
              "Retail Price Per Gallon": "APU000074714",
              "Retail Sales (In Millions)": "RSGASS",
             "CPI": "CPIAUCSL"}
start = datetime.datetime(2000, 1, 1)
end = datetime.datetime.today()
data = gather_data(data_codes = data_codes,
                  start = start,
                  end = end,
                  freq = "M")
```

```python
data["Real Price Per Barrel"] = (
    data["Price Per Barrel"] * 301.808).div(data["CPI"])
data["Real Retail Price Per Gallon"] = (
    data["Retail Price Per Gallon"] * 301.808).div(data["CPI"])
data["Retail Sales (In Billions)"] = data["Retail Sales (In Millions)"] / 1000
data["Real Retail Sales (In Billions)"] = (
    data["Retail Sales (In Billions)"] * 301.808).div(data["CPI"])
```

```python
data_real = data[["Real Price Per Barrel",
                "Real Retail Price Per Gallon",
                "Real Retail Sales (In Billions)"]]
data_real = data_real.round(2)
data_real
```

```python
def plot_line(data, secondary_y, x1, x2, title):
    plt.rcParams.update({"font.size": 20})
    plot_data = data
    fig, ax = plt.subplots(figsize = (20, 12))
    plot_data.plot(secondary_y = secondary_y,
                   linewidth = 4, ax = ax)
    ax.axvspan(x1, x2, color = "k", linewidth = 0, alpha = 0.15)
    ax.set_title(title)
```

## Why Does Price of Oil Change?
### Jacob Beckmann

Major fluctuations in gas prices have always been somewhat of a mystery to me. Slight changes in the price of crude oil and the price of gasoline at the pump, whether they are up or down, are normal and can be seen even daily, but there have been quite a few major events since the year 2000 that have caused much more substantial dips or jumps in the price of crude oil and eventually the price of gasoline at the pump. Events that will be examined further within this writing that have been known to affect the price of crude oil are the September 11, 2001, terrorist attacks, the 2008 global financial crisis, the 2020 lockdown orders as a result of the COVID-19 pandemic, and, most recently, the war between Russia and Ukraine. What is it about these events that are causing these changes in the price of crude oil? This is a question I’m hoping to find answers to through research and the examination of relevant datasets.

### September 11, 2001

On September 11, 2001, the United States experienced a terrorist attack that is referred to today simply as 9/11. While this event left lasting effects on the United States’ spirits and has forever changed travel security, it also had large impacts on the U.S. economy. These impacts on the economy might have been shorter-lived when compared to other events to be discussed in this paper, but they are still worthy of note. In September 2001, crude oil prices were sitting at an average of \\$43.41 per barrel, and consumers were purchasing gas at \\$2.59 per gallon. After the attack on the 11th, it took only two months for crude oil prices to drop to a low of \\$31.82 per barrel, and gasoline prices to \\$1.92 per gallon. Something worthy of note is that in the last 24 years, this has been the lowest price gasoline has been per gallon, while crude oil reached its lowest price per barrel during the COVID-19 pandemic. A reason for the price of crude oil dropping can be attributed to simple supply and demand factors. After the attack, travel by flight greatly declined temporarily due to individuals feeling apprehensive about air travel. These decreases in air travel lowered demand for oil and subsequently lowered the price per barrel of crude oil.

### 2008 Global Financial Crisis

From 2004 to mid-2008, the price of oil rose to historic heights, only to fall drastically in the last quarter of 2008. This massive drop in oil prices essentially erased all price growth over the preceding four and a half years. The steepest price increase, seen from January 2007 to July 2008, was a difficult time for all economies. When prices began dropping quickly in August of 2008, it was a welcome sight for consumers; however, the cause of this decline was not as such. The global financial crisis hit its hardest beginning in the last quarter of 2008, hence the observed decline in oil prices. In mid-2008, crude oil prices reached a record-breaking high of around \\$183 per barrel, and in the span of just four months, they drastically fell to about \\$57 per barrel. This is a massive decrease of 69%. The price per gallon seen at the pump during this time period fell from \\$5.64 in July 2008 to the lowest \\$2.41 in December 2008, a 57% decrease. This severe worldwide recession from 2008–2009 drastically reduced economic activity. This reduction in economic activity caused a large decrease in demand for crude oil and other petroleum products; therefore, the prices of associated products experienced lower prices until economies began to recover. The price of crude oil per barrel just before the crash was at an all-time high and wouldn’t reach a similar level until April 2011, with a price of \\$166 per barrel, and again in June 2022, with a price of \\$126 per barrel. Both instances are still just shy of the price per barrel pre-market crash by about \\$17 and \\$57. Something of interest: the price of crude oil in the past 24 years has never reached a higher value than it had just before the 2008 market crash.

### 2015 Dip

From April 2011 to June 2014, the price per barrel of crude oil was hovering at an average of about \\$145, give or take about \\$15; however, these prices would soon change dramatically. In June 2014, the price of crude oil was sitting at \\$142.23 per barrel. Over the course of the next 7 months, crude oil would drop to a low of \\$61.40. This was a drop of about 57%. At the pump, prices dropped to \\$2.17 per gallon from \\$4.70 per gallon. This was a drop of about 54%. In the last quarter of 2014, there was an oversupply of crude oil on the world market. This large surplus overshadowed demand and caused an acceleration in the drop in overall crude oil prices, hence the 57% price drop. One contributor to the surplus in oil was U.S. production. Since a peak in 2005, U.S. dependency on outside-sourced petroleum has declined. In 2013, U.S. oil production rose to the highest it had been in 24 years. This strong production of oil from the U.S. continued into 2014. Another factor that led to an increase in global oil supply was an increase in production from OPEC. In November of 2014, OPEC met to discuss cutting production levels due to declining prices. A decision was ultimately made to keep production steady at current levels. This decision surprised petroleum markets, further pushing prices down.

```python
plot_line(data_real[["Real Retail Price Per Gallon", "Real Price Per Barrel"]],
         "Real Retail Price Per Gallon", 
         "2014-05", "2015-02", "2015 Dip")
```

### COVID-19

In late 2019 or early 2020, a virus known as COVID-19 began spreading across the globe. During the initial stages of the virus’s growth, no one knew just how drastically our lives would be affected in the coming months. Jump ahead to March 2020, and the first states in the U.S. will enact mandated lockdown orders to prevent the spread of the virus, with many countries in Europe and Asia already in lockdown. This worldwide shutdown had a major impact on the day-to-day lives of nearly everyone. As well as changing the way we all lived personally, the stay-at-home orders drastically affected the world economy. More specifically, these lockdowns had a massive impact on the price of crude oil and, subsequently, the price observed by consumers at the pump. In December of 2019, shortly before COVID began impacting the world, crude oil prices were at a high of about \\$78 per barrel, or an average of \\$3.02 per gallon at the pump. After the virus hit hard and lockdowns had been put into place, the price of crude oil per barrel dropped to a low of about \\$22, or an average of \\$2.21 per gallon seen by consumers in April 2020. This was the lowest crude oil price per barrel experienced since February of 1999. It wouldn’t be until April 2021, over a full year after U.S. lockdowns began, that crude oil prices returned to their original value pre-shutdown. The cause of this decrease in crude oil prices is much simpler when compared to the causes of other drops. This is a textbook case of supply and demand rules. In December of 2019, nearly \\$51.33 billion was purchased by consumers. After strict lockdowns were put in place, this value dropped to its lowest value of \\$31.76 billion in April 2020, a decrease of nearly 40 percent. This large fall in spending due to lockdown orders caused a sudden decrease in demand for oil. In turn, oil producers were left with a large surplus of crude oil and had no choice but to drastically cut the price per barrel, thus resulting in a subsequent 27% decrease in the price per gallon at the pump.

### May 2022 spike

In June 2022, the price of crude oil was at a peak of \\$125.66 per barrel, and retail prices were at a price of \\$5.18 per gallon. These were the highest prices seen since just before the crash in prices in June 2014. Prices continuously rose to this point from an all-time low in March 2020, when crude oil was priced at an average of \\$32 per barrel. Over the course of a little over two years, the price of crude oil rose by nearly 393\% to reach the level it was at in May 2022. There were several contributing factors that led to this steep increase in the price of oil. One factor leading to this increase in price was rising inflation. Following the pandemic caused by COVID-19, aid for individuals and the economy was distributed in the form of stimulus checks. Multiple times over the course of the pandemic, the U.S. government distributed stimulus checks to individuals to encourage spending. This additional insertion of cash into the economy caused increasing rates of inflation that were likely attributed to higher prices of crude oil and consequently higher retail prices seen at the gas pump. Additionally, this increase in spending power consumers received via stimulus checks was a cause of rising gas prices due to the decrease in value of the U.S. dollar and the increase in demand the checks created. Another contributor to the increase in crude oil prices at the time was the war between Russia and Ukraine, more specifically, the invasion of Ukraine by Russia. While the war between Russia and Ukraine first began in February 2014, Russia recently invaded Ukraine in February 2022. Only three months after the invasion did the price of crude oil reach its peak. Russia is a large supplier of oil to the world market, so not long after the invasion, in an attempt to punish their economy for the invasion of Ukraine, the United States, along with other countries, began placing economic sanctions on Russia. With Russia no longer being able to supply oil to the U.S. due to the sanctions, demand increased drastically for other importers. This increase in demand contributed to the increase in the price of crude oil. Rising inflation in the United States due to COVID-19 stimulus checks combined with the shifting of crude oil supply and demand from economic sanctions placed on Russia contributed to the large spike in gas prices seen in June 2022. Since then, prices have begun falling to more stable levels, with the price of crude oil sitting at \\$37.43 per barrel and retail gas prices at an average of \\$3.55 per gallon.

### Conclusions

In conclusion, large fluctuations in crude oil and gas prices aren’t as mysterious as they might seem. When the market is experiencing a major shift, up or down, in crude oil prices, there will almost certainly be some external event driving these prices. Sometimes it’s as simple as basic supply and demand factors, like when individuals were flying less after the 9/11 terrorist attacks or when people were staying home for extended periods of time due to COVID-19 lockdown restrictions. Both of these events greatly reduced demand for oil, driving prices down. Other times, the cause might be a bit more complex. For example, the large dip in 2015 was due to more complicated supply reasoning; the U.S. was producing much more oil than in years past, and OPEC underestimated the effects of declining price levels. The combination of these two things caused a major surplus and dropped the price of crude oil greatly. Another more complicated event was the major spike in prices seen in May 2022. This spike happened because of several events, which included inflation due to U.S. stimulus checks and economic sanctions placed on Russia as a result of the Russian-Ukraine war. So, while these events may seem confusing at first, they all seem to make a bit more sense once you take a closer look.
