# Project Proposal: Examining the Cause of Gas Price Fluctuation

Major fluctuations in gas prices have always been somewhat of a mystery to me. Slight changes in prices at the pump are very normal and can be seen every day, but there have been quite a few events in the past decade that have caused major dips or major jumps in the price per gallon. These larger dips and jumps are what I’ve always been curious about. What events are happening that are causing these changes? Why do these events have an effect on prices? There are also occurrences that were believed to have an effect on gas prices but were found to have no impact on the price seen at the pump. What events are falsely believed to effect prices? Why do people believe these events are impacting prices? These questions are ones that I will hopefully come closer to answering within this paper.

The first major dip that can be seen is in December 2008. This dip is likely correlated with the 2008 Financial Crisis. This crisis was the most serious financial crisis experienced since the Great Depression. This recession was caused by the bursting of a housing bubble, along with other factors. In December 2008, average retail gas prices fell to a low of 1.75 per gallon. Only 5 months prior, in July of 2008, gas prices were at a high of 4.11 per gallon. That is a major reduction of about 57 percent.

Another large drop occurred in January 2015 and again in February 2016. The first drop in 2015 was primarily due to supply factors. Another drop was seen in 2016 due to similar reasons. This is something that I will look into further to because simply saying prices dropped due to “supply factors” isn’t very explanatory.

A more recent decline was seen in April of 2020. This decline was caused by the covid-19 virus, or more directly by the lack of travel that occurred because of the lockdowns associated with the beginning of the pandemic. When covid began spreading in the U.S., states quickly began putting stay-at-home orders in place. Because so many people were now staying in their homes, there was a very large reduction in travel. This reduction in travel caused a decrease in demand for gas which in turn resulted in the lower price per gallon gas. At its lowest, gas reached an average price of $1.93 per gallon. Over the course of the next year, as states began removing their stay-at-home orders, gas prices began rising back to their original price pre-pandemic.

An ongoing event affecting the price of gasoline is the war between Russia and Ukraine. At its peak, the price per gallon of gasoline rose to $5.03. While the war first began in February of 2014, the recent invasion of Ukraine by Russia in February 2022 had major effects on the world market. After the invasion of Ukraine, the United States, along with other countries, banned imports of Russian energy. Because Russia is a large supplier of crude oil, after the ban on Russian energy, countries had to resort to getting their oil supplies from other places. This caused a mass increase in oil demand for some places which raised up the price of oil per barrel. This in turn caused in increase in prices seen at the pump.

In January 2021, the Keystone Pipeline was shut down and would not be built. This was a source of a lot of controversy for a short while. There were many people who believed the rising gas prices at that time was due to the cancelation of the pipeline. In reality, when the pipeline was cancelled it wasn’t yet operational, therefore it had no effect yet on oil supply. It is likely that the rising price of gas at the time was due to covid lockdown restrictions being removed. As stay-at-home orders were ending, people were beginning to travel more. This rise in travel caused an increase in demand for gasoline which would then raise the price of gas.

I intend to further examine the ideas listed above to gain more insight behind the reasons we experience changing prices in gasoline.
